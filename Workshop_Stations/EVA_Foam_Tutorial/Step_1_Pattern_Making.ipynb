# Step 1 – Pattern Making

Before cutting any EVA foam, create paper templates for the glasses case. Using paper first allows you to check the size, shape, and fit before committing to the final material. The finished template should resemble the shape shown in Figure 1.

### 1. Fold the paper
Fold a sheet of paper in half to create the centre fold line of the case.

### 2. Mark the hinge area
Along the fold line, draw two short parallel lines approximately 2 cm long to define the hinge area between the front and back of the case.

### 3. Size the rectangular section
Ensure that the rectangular section is slightly shorter than the width of the folded glasses. This allows it to fit between the lenses and the folded temples (legs), helping to hold the glasses securely inside the case.

### 4. Trace the glasses outline
Place the folded glasses against the fold line and trace the outline of one half, leaving 1.5–2 cm of clearance around the glasses for adhesives.

### 5. Cut and unfold
Cut out the template while the paper is still folded, then unfold it to create a symmetrical case.

### 6. Create the strap templates
Create separate templates for the retaining strap and the hanging strap while the paper is still folded.

### 7. Strap dimensions
Make the retaining strap approximately 4 cm × 2 cm and the hanging strap approximately 8 cm × 2 cm. These dimensions are only a starting point and can be adjusted during the fit test.

Test all templates on the glasses and adjust as needed before transferring them onto the EVA foam.
